```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
x=[1,2,3,4,5,6,7]
y=[20,40,30,40,30,50,30]

df = pd.DataFrame({"Days":x,"No of people":y})
df.head(7)
```

```python
sns.lineplot(x="Days",y="No of people",data=df)
plt.show()
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.lineplot(x="day" ,y="total_bill" ,data = df,hue = "sex",style="time",palette="")
plt.xlabel("total bill")
plt.ylabel("day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x="who",y="fare" , data=df ,color="magenta")
plt.show()
```

```python
sns.barplot(x="who",y="fare" , data=df ,hue="sex",palette="spring",saturation=10)
plt.show()
```

```python
sns.barplot(x="fare",y="who" , data=df ,hue="sex",palette="spring",saturation=10)
plt.show()
```

```python
sns.barplot(x="fare",y="who" , data=df ,ci=False)
plt.show()
```

```python
from numpy import median
sns.barplot(x="fare",y="who",data=df,estimator="median")
plt.show()
```

```python
sns.barplot(x="fare",y="who" , data=df ,facecolor=[1,1,1,0],edgecolor="0.2")
plt.show()
```

```python
df=sns.load_dataset("tips")
df.head()
```
